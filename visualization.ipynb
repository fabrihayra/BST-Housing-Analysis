```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp = [300.2, 543.5, 1076.8, 3498.3, 7985.9, 10289.7, 11343.0, 13487.2]
```

```python
from collections import Counter

# sample sales data: product names and quantities sold

transactions = [
    ("Laptop", 5),
    ("Headphones", 15),
    ("Mouse", 10),
    ("Laptop", 7),
    ("Monitor", 3),
    ("Headphones", 20),
    ("Keyboard", 8),
    ("Mouse", 12),
    ("Monitor", 5),
    ("Mouse", 6),
]

# aggregating sales data by product
product_sales = Counter()

# summing up the total quantity sold for each product
for product, quantity in transactions:
    product_sales[product] += quantity

# categorizing products by sales volume
sales_categories = {
    "Low Volume": [],
    "Medium Volume": [],
    "High Volume": [],
}

# attributing volumes to sales
for product, sales in product_sales.items():
    if sales < 10:
        sales_categories["Low Volume"].append(product)
    elif 10 <= sales <= 20:
        sales_categories["Medium Volume"].append(product)
    else:
        sales_categories["High Volume"].append(product)

# displaying results
print("Total Sales by Product:")
for product, sales in product_sales.items():
    print(f"{product}: {sales}")

print ("\nProducts by Sales Volume:")
for category, products in sales_categories.items():
    print(f"{category}: {products}")
```

```python
import matplotlib.pyplot as plt

products = list(product_sales.keys())
quantities = list(product_sales.values())

# total sales by product
plt.bar(products, quantities, color='pink', edgecolor='black')

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# combining products and quantities into tuples
sorted_data = sorted(zip(quantities, products))

# unziping the sorted list into sorted quantities and products
sorted_quantities, sorted_products = zip(*sorted_data)

# creating the bar chart with sorted data
plt.bar(sorted_products, sorted_quantities, color='pink', edgecolor='black')

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
